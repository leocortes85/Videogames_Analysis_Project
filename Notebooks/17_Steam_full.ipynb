```python
def data_summ(df, title=None):
    '''
    Function to provide detailed information about the dtype, null values,
    and outliers for each column in a DataFrame.

    Parameters:
    - df (pd.DataFrame): The DataFrame for which information is to be generated.
    - title (str, optional): Title to be used in the summary. If None, the title will be omitted.

    Returns:
    - df_info (pd.DataFrame): A DataFrame containing information about each column,
                              including data type, non-missing quantity, percentage of
                              missing values, missing quantity, and information about outliers.
    '''
    info_dict = {"Column": [], "Data_type": [], "No_miss_Qty": [], "%Missing": [], "Missing_Qty": []}

    for column in df.columns:
        info_dict["Column"].append(column)
        info_dict["Data_type"].append(df[column].apply(type).unique())
        info_dict["No_miss_Qty"].append(df[column].count())
        info_dict["%Missing"].append(round(df[column].isnull().sum() * 100 / len(df), 2))
        info_dict['Missing_Qty'].append(df[column].isnull().sum())

  
    df_info = pd.DataFrame(info_dict)

    if title:
        print(f"{title} Summary")
        print("\nTotal rows: ", len(df))
        print("\nTotal full null rows: ", df.isna().all(axis=1).sum())

    
    return df_info
```

```python
# Hacer el merge usando ambas columnas 'Item_id' y 'Item_name'
df_full = pd.merge(df1, df2, on=['Item_id', 'Item_name', 'Genres', 'Release', 'Developer', 'Publisher' ], how='outer')

# Mostrar el resultado
df_full
```

```python
# Crear la nueva columna 'Price' tomando el valor más alto entre 'Price_x' y 'Price_y'
df_full['Price'] = df_full[['Price_x', 'Price_y']].apply(
    lambda x: max(x) if pd.notna(max(x)) else 0, axis=1
)

# Si deseas, puedes eliminar las columnas anteriores 'Price_x' y 'Price_y'
df_full = df_full.drop(columns=['Price_x', 'Price_y'])

# Mostrar el resultado
df_full
```

```python
# Guardar el orden original de las columnas
original_column_order = df_full.columns

# Contar el número de valores no nulos en cada fila
df_full['non_null_count'] = df_full.notnull().sum(axis=1)

# Encontrar la fila con el máximo número de valores no nulos para cada Item_id
idx = df_full.groupby('Item_id')['non_null_count'].idxmax()

# Crear un DataFrame con las filas que tienen el máximo número de valores no nulos
max_info_df_full = df_full.loc[idx].drop(columns='non_null_count')

# Actualizar columnas numéricas con el valor más alto para cada Item_id
num_cols = df_full.select_dtypes(include=np.number).columns
max_values_df_full = df_full.groupby('Item_id')[num_cols].max()

# Combinar DataFrames para rellenar los valores vacíos y actualizar columnas numéricas
df_full_filled = df_full.set_index('Item_id').combine_first(max_info_df_full.set_index('Item_id')).reset_index()
df_full_filled[num_cols] = df_full_filled.groupby('Item_id')[num_cols].transform('max')

# Opcional: Eliminar la columna 'non_null_count' si ya no la necesitas
df_full_filled = df_full_filled.drop(columns='non_null_count')

# Restaurar el orden original de las columnas
df_full_filled = df_full_filled[original_column_order]

# Asignar el DataFrame procesado al original
df_full = df_full_filled
```
