```python

# Mount external lakehouse  "Silver" into specified local path.
# Note: NotebookUtils is only supported on runtime v1.2 and above. If you are using runtime v1.1, please use mssparkutils instead.
notebookutils.fs.mount("abfss://8c7d2edc-a6a8-4f1a-bef1-c8b14449cfd3@onelake.dfs.fabric.microsoft.com/03b0b8ae-98d8-43f5-a727-9cd976a6117b", "/lakehouse/Silver")

# Load data into pandas DataFrame from notebookutils.fs.getMountPath('/lakehouse/Silver/Files/Cleaned_Data/steam_full.parquet')
df_steam = pd.read_parquet(notebookutils.fs.getMountPath('/lakehouse/Silver/Files/Cleaned_Data/steam_full.parquet'))
display(df_steam)
```

```python
# Generar un valor aleatorio entre 5 y 10
def random_multiplier():
    return np.random.uniform(5, 10)

# Crear la columna 'Global_Sales' con el ajuste para 'Price' igual a 0
df_steam['Global_Sales'] = df_steam.apply(
    lambda row: (row['Owners'] * row['Price'] if row['Price'] > 0 else 
                 row['Owners'] * random_multiplier()) / 1_000_000,
    axis=1
)
# Redondear 'Global_Sales' a dos decimales
df_steam['Global_Sales'] = df_steam['Global_Sales'].round(2)
df_steam
```

```python
# Añadir columnas faltantes a df_steam
for col in df_console.columns:
    if col not in df_steam.columns:
        df_steam[col] = pd.NA

# Añadir columnas faltantes a df_console
for col in df_steam.columns:
    if col not in df_console.columns:
        df_console[col] = pd.NA

# Reordenar las columnas para que ambos DataFrames tengan el mismo orden
df_console = df_console[df_steam.columns]

# Concatenar los DataFrames verticalmente
df_combined = pd.concat([df_steam, df_console], ignore_index=True)

games_full = df_combined

games_full
```

```python
# Borrar las columnas especificadas
columns_to_drop = ['positive_ratings', 'negative_ratings', 'Average_playtime', 'Owners', 'Price', 'Score',	'NA_Sales',	'EU_Sales',	'JP_Sales',	'Other_Sales']
games_full = games_full.drop(columns=columns_to_drop)
```

```python
games_full
```

```python
# Separar los DataFrames antes y después de la concatenación
df1 = games_full[games_full['Item_id'].notna()]  # DataFrame con Item_id no nulos
df2 = games_full[games_full['Item_id'].isna()]   # DataFrame con Item_id nulos

# Obtener los valores únicos existentes en 'Item_id' de df1
existing_ids = set(df1['Item_id'])

# Generar una lista de nuevos valores únicos que no estén en existing_ids
max_id = 1
while max_id in existing_ids:
    max_id += 1

# Crear una lista de nuevos valores
new_ids = []
while len(new_ids) < len(df2):
    if max_id not in existing_ids:
        new_ids.append(max_id)
    max_id += 1

# Asignar los nuevos valores únicos a df2 en orden ascendente
df2['Item_id'] = new_ids

# Concatenar de nuevo los DataFrames
df_final = pd.concat([df1, df2], ignore_index=True)

# Ordenar el DataFrame final por 'Item_id' en orden ascendente
df_final = df_final.sort_values(by='Item_id').reset_index(drop=True)

# Mostrar el DataFrame final
df_final
```

```python
df_final['Item_id'] = pd.to_numeric(df_final['Item_id'], errors='coerce').astype('Int64')
df_final['Release'] = pd.to_numeric(df_final['Release'], errors='coerce').astype('Int64')
```

```python
archivo_parquet = '/lakehouse/default/Files/Final_data/games_full.parquet'

# Export to a parquet file.
df_final.to_parquet(archivo_parquet, index=False)
```
