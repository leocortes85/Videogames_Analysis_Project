```python
import pandas as pd
import numpy as np
import pyarrow.parquet as pq
import pyarrow as pa
```

```python
# Load data into pandas DataFrame from "/lakehouse/default/Files/Final_data/games_full.parquet"
df = pd.read_parquet("/lakehouse/default/Files/Final_data/games_full.parquet")
df.head()
```

```python
#show unique values in Platform column
unique_plat = df['Platform'].unique()
print(unique_plat)
```

```python
platform = df[["Platform"]].drop_duplicates().reset_index(drop=True)
platform['Platform_id'] = platform.index + 1
order = ['Platform_id', 'Platform']
platform = platform[order]
platform
```

```python
genres = df[["Genres"]].drop_duplicates().reset_index(drop=True)
genres['Genres_id'] = genres.index + 1
order = ['Genres_id', 'Genres']
genres = genres[order]
genres
```

```python
games = df[['Item_id', 'Item_name']].drop_duplicates().reset_index(drop=True)
games
```

```python
dev = df[["Developer"]].drop_duplicates().reset_index(drop=True)
dev['Developer_id'] = dev.index + 1
order = ['Developer_id', 'Developer']
dev = dev[order]
dev
```

```python
publish = df[["Publisher"]].drop_duplicates().reset_index(drop=True)
publish['Publisher_id'] = publish.index + 1
order = ['Publisher_id', 'Publisher']
publish = publish[order]
publish
```

```python
rating = df[["Rating"]].drop_duplicates().reset_index(drop=True)
rating['Rating_id'] = rating.index + 1
order = ['Rating_id', 'Rating']
rating = rating[order]
rating
```

```python
games_fact = df[['Item_id',	'Platform',	'Genres',	'Release',	'Developer',	'Publisher',	'Rating',	'Global_Sales']]
games_fact = games_fact.merge(platform, on='Platform', how='left')
games_fact = games_fact.merge(genres, on='Genres', how='left')
games_fact = games_fact.merge(dev, on='Developer', how='left')
games_fact = games_fact.merge(publish, on='Publisher', how='left')
games_fact = games_fact.merge(rating, on='Rating', how='left')
columns_to_drop = ['Developer', 'Platform', 'Genres', 'Rating', 'Publisher']
games_fact = games_fact.drop(columns_to_drop, axis=1)
games_fact['Release'] = pd.to_datetime(games_fact['Release'], format='%Y')
column_order = ['Item_id', 'Release', 'Platform_id', 'Genres_id', 'Developer_id', 'Publisher_id', 'Rating_id', 'Global_Sales']
games_fact = games_fact[column_order]
games_fact
```

```python
def save_to_pq(dfs, names):

    """
    Saves a DataFrame to a parquet file in a specific directory.
        Parameters:
        - df: DataFrame to save.
        - file_name: Name of the parquet file.
    """
    for df, name in zip(dfs, names):
        archivo = f'/lakehouse/default/Files/Final_data/Analytics/{name}.parquet'
        pq.write_table(pa.Table.from_pandas(df), archivo)
        print(f"DataFrame '{name}' save as '{archivo}'")
```

```python
dfs = [games_fact, games, platform, genres, dev, publish, rating]
names = ['games_fact', 'games', 'platforms', 'genres', 'devs', 'publish', 'rating']

save_to_pq(dfs, names)
```
